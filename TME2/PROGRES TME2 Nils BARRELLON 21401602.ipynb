> J'aimerais que tu me donnes un script python qui : 1) d√©compresse deux fichiers intitul√©s recipes.txt.gz et ingredients.txt.gz pour en extraire les 2 fichiers ingredients.txt et recipes.txt. 2) le code ouvre les deux fichiers .txt et les stocke dans deux dictionnaires. Un dictionnaire nomm√© ingr√©dients dont les cl√©s sont des ingr√©diens et la valeur la cat√©gorie √† laquelle ils appartiennent Vegetables,Fruits... Un dictionnaire nomm√© recipe dont les cl√©s sont le num√©ro de la recette et les valeurs une liste des ingr√©dients qui la compose. 3) parcourt les recettes du dictionnaire recette et, pour chaque recette, parcourt les ingr√©dients qui la composent. Le code renvoit donc cette liste et sa taille. 4) Ce code devra √™tre compos√© de deux fonctions : lecture_fichier, verification_ingr√©dient.

```python
import gzip

def decompresser_fichiers(chemin_recipes_gz, chemin_ingredients_gz):
    # D√©compresser les fichiers .gz
    with gzip.open(chemin_recipes_gz, 'rt', encoding='utf-8') as f_in, open('recipes.txt', 'w', encoding='utf-8') as f_out:
        f_out.write(f_in.read())

    with gzip.open(chemin_ingredients_gz, 'rt', encoding='utf-8') as f_in, open('ingredients.txt', 'w', encoding='utf-8') as f_out:
        f_out.write(f_in.read())

def lecture_fichier(chemin_ingredients, chemin_recettes):
    # Dictionnaire pour stocker les ingr√©dients et leurs cat√©gories
    ingredients = {}

    # Lire le fichier des ingr√©dients
    with open(chemin_ingredients, 'r', encoding='utf-8') as f:
        for line in f:
            if ':' in line:
                categorie, liste_ingredients = line.split(':', 1)
                categorie = categorie.strip()
                for ingredient in liste_ingredients.split(','):
                    ingredient = ingredient.strip()
                    ingredients[ingredient] = categorie

    # Dictionnaire pour stocker les recettes
    recettes = {}

    # Lire le fichier des recettes
    with open(chemin_recettes, 'r', encoding='utf-8') as f:
        for line in f:
            if 'Recipe #' in line:
                numero_recette, liste_ingredients = line.split(':', 1)
                numero_recette = numero_recette.split('#')[1].strip()
                ingredients_recette = [ing.strip() for ing in liste_ingredients.split(',')]
                recettes[numero_recette] = ingredients_recette

    return ingredients, recettes

def verification_ingredient(ingredients, recettes):
    # Liste pour stocker les recettes sans Proteins ni Dairy
    recettes_sans_proteins_sans_dairy = []

    # Parcourir chaque recette
    for numero_recette, liste_ingredients in recettes.items():
        recette_valide = True

        # Parcourir chaque ingr√©dient de la recette
        for ingredient in liste_ingredients:
            if ingredient in ingredients and ingredients[ingredient] in ['Proteins', 'Dairy']:
                recette_valide = False
                break

        # Si la recette est valide, l'ajouter √† la liste
        if recette_valide:
            recettes_sans_proteins_sans_dairy.append(numero_recette)

    return recettes_sans_proteins_sans_dairy, len(recettes_sans_proteins_sans_dairy)

# Exemple d'utilisation
chemin_recipes_gz = 'recipes.txt.gz'
chemin_ingredients_gz = 'ingredients.txt.gz'

# √âtape 1 : D√©compresser les fichiers
decompresser_fichiers(chemin_recipes_gz, chemin_ingredients_gz)

# √âtape 2 : Lire les fichiers et stocker dans des dictionnaires
ingredients, recettes = lecture_fichier('ingredients.txt', 'recipes.txt')

# √âtape 3 : V√©rifier les recettes
recettes_sans_proteins_sans_dairy, nombre_recettes = verification_ingredient(ingredients, recettes)

print(f"Nombre de recettes : {nombre_recettes}")
print(f"Recettes sans Proteins ni Dairy : {recettes_sans_proteins_sans_dairy}")



```

La fonction affiche alors la recette "pr√©f√©r√©e" et sa liste d'ingr√©dients (les pond√©rations seront indiqu√©es entre parenth√®ses).
```python
import random

def my_favourite_recipe(ingredients, recettes):
    # Cr√©er un dictionnaire de pond√©rations al√©atoires pour chaque ingr√©dient
    my_favourite = {ingredient: random.randint(1, 15) for ingredient in ingredients.keys()}

    # Liste pour stocker les scores des recettes
    scores_recettes = []

    # Calculer le score de chaque recette
    for numero_recette, liste_ingredients in recettes.items():
        score = 0
        for ingredient in liste_ingredients:
            if ingredient in my_favourite:
                score += my_favourite[ingredient]
        scores_recettes.append((numero_recette, score, liste_ingredients))

    # Trouver le score maximum
    max_score = max(score for _, score, _ in scores_recettes)

    # Trouver toutes les recettes avec le score maximum
    recettes_preferees = [(numero_recette, score, liste_ingredients)
                          for numero_recette, score, liste_ingredients in scores_recettes
                          if score == max_score]

    # Afficher les recettes pr√©f√©r√©es
    print("Recette(s) pr√©f√©r√©e(s) :")
    for numero_recette, score, liste_ingredients in recettes_preferees:
        print(f"Recette #{numero_recette} (Score: {score})")
        print("Ingr√©dients (pond√©ration) :")
        for ingredient in liste_ingredients:
            if ingredient in my_favourite:
                print(f"- {ingredient} ({my_favourite[ingredient]})")
        print()

my_favourite_recipe(ingredients, recettes)        
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# G√©n√©ration de donn√©es al√©atoires
# ----------------------------
np.random.seed(0)  # pour reproductibilit√©
n_points = 50
points = np.random.rand(n_points, 2)  # valeurs dans [0, 1]
# On peut multiplier par 100 pour des valeurs plus lisibles :
points *= 100

# ----------------------------
# Fonction de calcul du front de Pareto (maximisation)
# ----------------------------
def pareto_front(points):
    """
    Renvoie les points non domin√©s (front de Pareto)
    pour un probl√®me de maximisation √† 2 crit√®res.
    """
    pareto = []
    for i, p in enumerate(points):
        # Un point est domin√© si un autre est >= sur les deux crit√®res
        # et strictement > sur au moins un
        dominated = False
        for j, q in enumerate(points):
            if i != j and np.all(q >= p) and np.any(q > p):
                dominated = True
                break
        if not dominated:
            pareto.append(p)
    return np.array(pareto)

pareto = pareto_front(points)

# ----------------------------
# Trac√©
# ----------------------------
plt.figure(figsize=(7, 6))
plt.scatter(points[:, 0], points[:, 1], label="Solutions", color="gray", alpha=0.6)
plt.scatter(pareto[:, 0], pareto[:, 1], color="red", label="Front de Pareto", s=70)

# Pour visualiser le front : on relie les points tri√©s
pareto_sorted = pareto[np.argsort(pareto[:, 0])]
plt.plot(pareto_sorted[:, 0], pareto_sorted[:, 1], color="red", linestyle="--")

plt.xlabel("Crit√®re 1 (√† maximiser)")
plt.ylabel("Crit√®re 2 (√† maximiser)")
plt.title("Front de Pareto (maximisation des deux crit√®res)")
plt.legend()
plt.grid(True)
plt.show()
```

ChatGPT m'a gentiment expliqu√© que cette fonction avait une complexit√© dans le pire des cas en O(n<sup>2</sup>) du fait des deux boucles imbriqu√©es. Mon prompt :
> quelle est la complexit√© dans le pire des cas de cette fonction ?

La premi√®re courbe faisait appara√Ætre quelques valeurs aberrantes. ChatGPT m'a propos√© de me donner une deuxi√®me fonction qui rep√™te dix fois une mesure pour un n donn√© afin de lisser les donn√©es. Le r√©sultat est "plus joli". C'est ce deuxi√®me code que je colle ci-dessous. Mon prompt :
>Je veux √©valuer de fa√ßon empirique la complexit√© de cette fonction. Je souhaite donc que tu me donnes le code d'une fonction nommee verif_complexite qui re√ßoit une liste de points de taille n et renvoie le temps necessaire pour que la fonction pareto_front fasse son travail sur ces points. Dans le programme principal, tu cr√©eras une liste de 90 points dans l'espace 100,100, tu feras une boucle de n = 10 √† n= 100, tu enverras la la fonction verif_complexite un slice de la liste points de 0 √† n pour remplir une liste du temps en fonction de la taille de l'entr√©e n. Enfin, tu traceras le graphe t=f(n). Sur le m√™me graphique, tu afficheras la courbe th√©orique t=f(n¬≤)

```python
import time

# ----------------------------
# Fonction Pareto √† √©tudier
# ----------------------------
def pareto_front(points):
    """
    Renvoie les points non domin√©s (front de Pareto)
    pour un probl√®me de maximisation √† 2 crit√®res.
    """
    pareto = []
    for i, p in enumerate(points):
        dominated = False
        for j, q in enumerate(points):
            if i != j and np.all(q >= p) and np.any(q > p):
                dominated = True
                break
        if not dominated:
            pareto.append(p)
    return np.array(pareto)

# ----------------------------
# Fonction de v√©rification de la complexit√©
# ----------------------------
def verif_complexite(points, n_repet=5):
    """
    Mesure le temps moyen d'ex√©cution de pareto_front sur la liste de points donn√©e.
    Effectue n_repet r√©p√©titions pour lisser les fluctuations.
    Retourne le temps moyen en secondes.
    """
    temps = []
    for _ in range(n_repet):
        debut = time.perf_counter()
        pareto_front(points)
        fin = time.perf_counter()
        temps.append(fin - debut)
    return np.mean(temps)

# ----------------------------
# Programme principal
# ----------------------------
np.random.seed(0)

# G√©n√©ration de 100 points al√©atoires dans [0,100]
points = np.random.rand(100, 2) * 100

# Listes pour stocker les r√©sultats
n_values = list(range(10, 101, 2))
t_values = []

# Mesure du temps pour chaque taille n (moyenn√© sur plusieurs r√©p√©titions)
for n in n_values:
    pts = points[:n]
    t = verif_complexite(pts, n_repet=10)
    t_values.append(t)

# ----------------------------
# Courbe th√©orique en O(n¬≤)
# ----------------------------
n2_values = np.array(n_values) ** 2
# Ajustement de l‚Äô√©chelle
coef = t_values[-1] / n2_values[-1]
t_theo = coef * n2_values

# ----------------------------
# Trac√© des r√©sultats
# ----------------------------
plt.figure(figsize=(8,6))
plt.plot(n_values, t_values, 'o-', label="Temps mesur√© (pareto_front)")
plt.plot(n_values, t_theo, '--', label="Courbe th√©orique ~ n¬≤")
plt.xlabel("Taille de l'entr√©e n", fontsize=12)
plt.ylabel("Temps d'ex√©cution moyen (s)", fontsize=12)
plt.title("V√©rification empirique de la complexit√© de pareto_front", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

De lui m√™me, chatGPT m'a sugg√©r√© un graphique log-log qui permet, par r√©gression lin√©aire, de d√©terminer le degr√© du polynome de la fonction O. En effet, sur la courbe ci-dessus, si on constate que les points exp√©rimentaux semblent "coller" √† la courbe t=f(n¬≤) cela reste flou.
```python
import numpy as np
import matplotlib.pyplot as plt
import time

# ----------------------------
# Fonction Pareto √† √©tudier
# ----------------------------
def pareto_front(points):
    pareto = []
    for i, p in enumerate(points):
        dominated = False
        for j, q in enumerate(points):
            if i != j and np.all(q >= p) and np.any(q > p):
                dominated = True
                break
        if not dominated:
            pareto.append(p)
    return np.array(pareto)

# ----------------------------
# Mesure du temps moyen
# ----------------------------
def verif_complexite(points, n_repet=5):
    temps = []
    for _ in range(n_repet):
        debut = time.perf_counter()
        pareto_front(points)
        fin = time.perf_counter()
        temps.append(fin - debut)
    return np.mean(temps)

# ----------------------------
# Programme principal
# ----------------------------
np.random.seed(0)
points = np.random.rand(1000, 2) * 100

n_values = list(range(10, 300, 5))
t_values = []

for n in n_values:
    pts = points[:n]
    t = verif_complexite(pts, n_repet=5)
    t_values.append(t)

# Courbe th√©orique n^2 normalis√©e
n2_values = np.array(n_values) ** 2
coef = t_values[-1] / n2_values[-1]
t_theo = coef * n2_values

# ----------------------------
# Calcul de la pente empirique
# ----------------------------
log_n = np.log(n_values)
log_t = np.log(t_values)

# R√©gression lin√©aire sur log-log : log(t) = a*log(n) + b
coeffs = np.polyfit(log_n, log_t, 1)
pente_empirique = coeffs[0]
intercepte = coeffs[1]

print(f"Pente empirique sur log-log : {pente_empirique:.2f}")
print(f"Intercept : {intercepte:.4f}")

# Trac√© avec droite de r√©gression
t_fit = np.exp(intercepte) * np.array(n_values) ** pente_empirique

plt.figure(figsize=(8,6))
plt.loglog(n_values, t_values, 'o', label="Temps mesur√©")
plt.loglog(n_values, t_fit, '-', label=f"R√©gression log-log : pente ‚âà {pente_empirique:.2f}")
plt.xlabel("Taille de l'entr√©e n (log)")
plt.ylabel("Temps d'ex√©cution moyen (s) (log)")
plt.title("Analyse log-log avec pente empirique")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import string

def dechiffre_cesar(texte: str, decalage: int) -> str:
    """
    D√©chiffre un texte chiffr√© par le chiffrement de C√©sar avec un d√©calage donn√©.
    """
    resultat = ""
    for c in texte:
        if 'a' <= c <= 'z':
            resultat += chr((ord(c) - ord('a') - decalage) % 26 + ord('a'))
        elif 'A' <= c <= 'Z':
            resultat += chr((ord(c) - ord('A') - decalage) % 26 + ord('A'))
        else:
            resultat += c
    return resultat


def score_frequence(texte: str) -> float:
    """
    Calcule un score de vraisemblance du texte bas√© sur la fr√©quence des lettres fran√ßaises.
    Plus le score est proche de la distribution fran√ßaise, plus le texte est probablement en fran√ßais.
    """
    # Fr√©quences approximatives des lettres en fran√ßais (en %)
    freq_fr = {
        'A': 7.6, 'B': 0.9, 'C': 3.3, 'D': 3.7, 'E': 14.7, 'F': 1.1, 'G': 1.0,
        'H': 0.7, 'I': 7.5, 'J': 0.6, 'K': 0.1, 'L': 5.5, 'M': 2.7, 'N': 7.2,
        'O': 5.8, 'P': 2.5, 'Q': 1.4, 'R': 6.5, 'S': 7.9, 'T': 7.1, 'U': 6.3,
        'V': 1.8, 'W': 0.0, 'X': 0.4, 'Y': 0.3, 'Z': 0.1
    }

    texte = texte.upper()
    lettres = [c for c in texte if c in string.ascii_uppercase]
    if not lettres:
        return 0

    total = len(lettres)
    freq_texte = {c: lettres.count(c) * 100 / total for c in string.ascii_uppercase}

    # Distance quadratique aux fr√©quences fran√ßaises
    score = -sum((freq_texte[c] - freq_fr[c]) ** 2 for c in string.ascii_uppercase)
    return score


def dechiffre_cesar_auto(texte: str) -> tuple[int, str]:
    """
    D√©tecte automatiquement le meilleur d√©calage pour un texte chiffr√© par C√©sar.
    Retourne (decalage_optimal, texte_dechiffre)
    """
    meilleurscore = float('-inf')
    meilleur_k = 0
    meilleur_txt = ""

    for k in range(26):
        candidate = dechiffre_cesar(texte, k)
        s = score_frequence(candidate)
        if s > meilleurscore:
            meilleurscore = s
            meilleur_k = k
            meilleur_txt = candidate

    return meilleur_k, meilleur_txt


# ----------------------------
# Exemple d'utilisation
# ----------------------------


def caesar(txt):
    decalage, texte_dechiffre = dechiffre_cesar_auto(txt)
    print(f"D√©calage d√©tect√© : {decalage}")
    print("-" * 80)
    return texte_dechiffre

txt = """Oh soxv fhoheuh gh fhv frghv hvw vdqv grxwh oh fkliiuh gh Fhvdu, 
xwlolvh sdu Mxohv gdqv vhv fruuhvsrqgdqfhv frqilghqwlhoohv. 
Lo frqvlvwh d ghfdohu fkdtxh ohwwuh gh wurlv udqjv, 
pdlv mh qh yrxv dssuhqgv ulhq sxlvtxh yrxv hwhv hq wudlq gh oluh fhwwh skudvh.
"""
caesar(txt)
```
