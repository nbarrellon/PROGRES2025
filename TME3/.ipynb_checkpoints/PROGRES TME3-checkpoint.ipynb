## IP adresses

- Use regular expressions to remove problematic zeros from an IPv4 address expressed as a 
string. (example: "216.08.094.196" should become "216.8.94.196", but "216.80.140.196" 
should remain "216.80.140.196"). 

```python
def normalize_ip(txt):
    cleaned_ip = re.sub(
        r'\b0+(\d)',  # Cherche un ou plusieurs zéros suivis d'un chiffre
        r'\1',        # Remplace par le chiffre seul
        txt
    )
    return cleaned_ip
```

```python
def normalize_ip(txt):
    decimaux = txt.split(".")
    ip_ok = ""
    for d in decimaux:
        if "0" in d:
            if d == "0":
                ip_ok+="0."
            elif d.startswith("00"):
                ip_ok += d[2:]+"."
            elif d.startswith("0"):
                ip_ok += d[1:]+"."
            else:
                ip_ok += d+"."
        else:
            ip_ok += d+"."    
            
    return ip_ok[:-1]
```

## Calendar convention

- Use regular expressions to transform a date from MM-DD-YYYY format to DD-MM-YYYY 
format. (example "11-06-2020" should become "06-11-2020").
- (*) Do the same thing using the `datetime` package.

Answer

```python
import re
def switch_md(txt):
    pattern = r'([0-9]{2})-([0-9]{2})-([0-9]{4})'
    assert re.fullmatch(pattern, txt) is not None,"La date n'est pas au format JJ-MM-AAAA"
    m = re.fullmatch(pattern, txt)
    jj,mm,aa = m.groups()
    return mm+"-"+jj+"-"+aa
```

```python
switch_md("11-06-2020")
```

```python
switch_md("25-12-2020")
```

```python
from datetime import datetime
def switch_md_bis(date_str):
    try:
        date_obj = datetime.strptime(date_str, "%d-%m-%Y").date()
        return date_obj.strftime("%m-%d-%Y")
    except ValueError:
        return None  # ou lever une exception si la chaîne n'est pas valide
```

```python
switch_md_bis("25-12-2020")
```

>Vérifier une adresse e-mail conforme aux RFC 822 (et ses mises à jour, notamment RFC 6532 qui autorise les caractères Unicode dans la partie locale) est un exercice complexe en raison de la grande variété de formats autorisés. Voici une expression régulière qui couvre la plupart des cas, y compris les adresses avec caractères Unicode (RFC 6532) et les formats classiques (RFC 822)

```python
import re

def is_valid_email(email):
    # RFC 822 + RFC 6532 (Unicode dans la partie locale)
    pattern = r"""
        ^                       # Début de la chaîne
        [\w\.\+\-]+             # Partie locale : lettres, chiffres, points, +, -, et certains caractères Unicode
        |                       # OU
        "                       # Partie locale entre guillemets (pour les caractères spéciaux)
        (?:[^"\\]|\\.)+         # Tout caractère sauf " et \, ou un caractère échappé
        "                       # Fin des guillemets
        @                       # @ obligatoire
        (?:                     # Début du domaine
            [\w\-]+             # Sous-domaine : lettres, chiffres, tirets
            (?:                 # Début d'un sous-domaine supplémentaire
                \.[\w\-]+       # Point suivi de lettres, chiffres, tirets
            )*                  # 0 ou plusieurs sous-domaines
            \.                  # Point avant le TLD
            [a-zA-Z]{2,}        # TLD : au moins 2 lettres
        |                       # OU
            \[                  # Domaine entre crochets (adresse IP)
            (?:                 # Début de l'adresse IP
                [0-9]{1,3}\.[0-9]{1,3}\.[0-9]{1,3}\.[0-9]{1,3}  # IPv4
            |                   # OU
                [a-fA-F0-9:]+   # IPv6
            )
            \]                  # Fin des crochets
        )
        $                       # Fin de la chaîne
    """
    return bool(re.fullmatch(pattern, email, re.VERBOSE))
```

```python
import re

def is_valid_email_bis(email):
    # Suppression des espaces en début et fin
    email = email.strip()
    # Regex simplifiée et corrigée
    pattern = r"""
        ^                       # Début de la chaîne
        [\w\.\+\-]+             # Partie locale : lettres, chiffres, points, +, -
        @                       # @ obligatoire
        [\w\-]+                 # Sous-domaine : lettres, chiffres, tirets
        (?:                     # Début des sous-domaines supplémentaires
            \.[\w\-]+           # Point suivi de lettres, chiffres, tirets
        )+                      # 1 ou plusieurs sous-domaines
        $                       # Fin de la chaîne
    """
    return bool(re.fullmatch(pattern, email, re.VERBOSE))
```

```python
for l in L:
    print(l," est valide ? ",is_valid_email_bis(l))
```

- Write a `display_cd` function that displays (i.e. `print`), for a CD: title, artist, country, company, year.
- Display all CDs.

```python
def display_cd(entry):
    infos = []
    for c in entry('CD'):
        titre = str(c('TITLE').pop())[7:-8]
        artiste = str(c('ARTIST').pop())[8:-9]
        pays = str(c("COUNTRY").pop())[9:-10]
        cie = str(c('COMPANY').pop())[9:-10]
        annee = str(c('YEAR').pop())[6:-7]
        infos.append({"titre":titre,"artiste":artiste,"pays":pays,"cie":cie,"annee":int(annee)})
    return infos
        
```

```python
# Regroup locations per movie
movies = [m for m in movies.values()]
movies
```

```python
from datetime import datetime
def display_movie(movie):
    ch = '"'+movie["titre"]+'",'+movie["realisateur"]+","+str(len(movie["locations"])) +" shooting(s),"
    d1 = movie["locations"][0][1].split("-")
    d2 = movie["locations"][0][2].split("-")
    earliest = datetime(int(d1[0]),int(d1[1]),int(d1[2]))
    latest = datetime(int(d2[0]),int(d2[1]),int(d2[2]))
    #on cherche la date la plus vieille (earliest) et la plus récente (latest)
    for loc in movie["locations"]:
        d1 = loc[1].split("-")
        d2 = loc[2].split("-")
        date_debut = datetime(int(d1[0]),int(d1[1]),int(d1[2]))
        date_fin = datetime(int(d2[0]),int(d2[1]),int(d2[2]))
        if date_debut<earliest:
            earliest = date_debut
        if date_fin>latest:
            latest = date_fin
    earliest = earliest.strftime("%m/%d/%Y").replace("/","-")
    latest = latest.strftime("%m/%d/%Y").replace("/","-")
    ch += "from "+earliest + " to " + latest
    return ch
```

```python
all_movie_displays = [display_movie(m) for m in movies]
print('\n'.join(all_movie_displays[:20]))
```

```python
import requests

def recuperer_contenu_page(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        return f"Erreur lors de la récupération de la page : {e}"

contenu_page = recuperer_contenu_page(url)
```

```python
from bs4 import BeautifulSoup

def extraire_noms_pays(html):
    soup = BeautifulSoup(html, 'html.parser')
    # Trouver le premier tableau de la page (celui qui contient les pays)
    tableau = soup.find('table', {'class': 'wikitable'})
    noms_pays = []
    if tableau:
        # Parcourir chaque ligne du tableau (sauf l'en-tête)
        for ligne in tableau.find_all('tr')[1:]:
            # La première colonne contient le nom du pays
            colonne_pays = ligne.find('td')
            if colonne_pays:
                nom_pays = colonne_pays.get_text(strip=True)
                noms_pays.append(nom_pays)
    return noms_pays

#html = recuperer_contenu_page(url)
countries = extraire_noms_pays(contenu_page)

```

```python
def extraire_donnees_pays(html):
    soup = BeautifulSoup(html, 'html.parser')
    tableau = soup.find('table', {'class': 'wikitable'})
    donnees_pays = []
    rang = 0
    if tableau:
        for ligne in tableau.find_all('tr')[1:]:  # On saute l'en-tête
            colonnes = ligne.find_all('td')
            if len(colonnes) >= 5:  # On vérifie qu'il y a assez de colonnes
                nom_pays = colonnes[0].get_text(strip=True).split('[')[0]  # On enlève les notes entre crochets
                densite = colonnes[2].get_text(strip=True)
                population = colonnes[3].get_text(strip=True)
                surface = colonnes[4].get_text(strip=True)
                donnees_pays.append({
                    "Rang": rang,
                    "Pays": nom_pays,
                    "Densité (hab/km²)": densite,
                    "Population": population,
                    "Surface (km²)": surface
                })
                rang += 1
    return donnees_pays

infos_countries = extraire_donnees_pays(contenu_page)[1:]
```

Gentiment, le chat me propose un code qui nécessite les bibliothèques FastAPI et HTTPException et uvicorn. Plusieurs aller-retours sont nécessaires pour que l'ensemble tourne. C'est finalement assez simple : on crée un serveur qui écoute sur le port 8090. Ce serveur possède plusieurs end-points qui redirige la requête vers le calcul demandé. Je constate juste qu'il faut relancer le kernel (puis le serveur) à chaque manipulation sur le code Python.

```python
from fastapi import FastAPI, HTTPException
import uvicorn
import nest_asyncio
import threading
import time

# Applique le patch pour permettre l'exécution asynchrone dans un notebook
nest_asyncio.apply()

app = FastAPI()

@app.get("/add/{a}/{b}")
async def addition(a: int, b: int):
    return {"resultat": a + b}

@app.get("/sous/{a}/{b}")
async def soustraction(a: int, b: int):
    return {"resultat": a - b}

@app.get("/mul/{a}/{b}") 
async def multiplication(a: int, b: int):
    return {"resultat": a * b}

@app.get("/div/{a}/{b}")
async def division(a: int, b: int):
    if b == 0:
        raise HTTPException(status_code=400, detail="Division par zéro impossible")
    return {"resultat": a / b}

@app.get("/mod/{a}/{b}")
async def modulo(a: int, b: int):
    return {"resultat": a % b}

# Fonction pour lancer Uvicorn dans un thread séparé
def run_server():
    uvicorn.run(app, host="0.0.0.0", port=8090)

# Lance le serveur dans un thread séparé
server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

# Attendre que le serveur soit prêt
time.sleep(3)
print("Serveur démarré sur http://localhost:8090")

```
